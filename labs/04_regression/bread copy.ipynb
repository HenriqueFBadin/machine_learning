```python
import numpy as np
```

```python
num_paes = 1000
```

```python
# Inventando paezinhos imaginários.

# Pesos: cerca de 50g, com desvio padrão de 3g.
pesos = 50 + 3*np.random.randn(num_paes, 1)

# Temperatura: cerca de 22 graus Celsius, com desvio padrão de 5 graus.
temps = 22 + 5*np.random.randn(num_paes, 1)

# Volume em ml: 3 ml por grama de massa, mais 4ml por grau acima de 22.
volume = 3*pesos + 4*(temps - 22) + 5*np.random.randn(num_paes, 1)
```

```python
X = np.c_[pesos, temps]
y = volume
```

```python
X.shape, y.shape
```

```python
# Vamos dividir em treino e teste.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# Aqui é onde o modelo aprende.
model.fit(X_train, y_train)
```

```python
model.intercept_, model.coef_
```

O modelo final é:

$$
\hat{y} = \overset{\theta_0}{-81.65} + \overset{\theta_1}{2.88} x_1 + \overset{\theta_2}{4.02} x_2
$$

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error

root_mean_squared_error(y_test, y_pred)
```
